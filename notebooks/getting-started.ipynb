## 2. Import Libraries

```python
from mplabml import Client
from getpass import getpass
import pandas as pd

print("✓ Libraries imported")
```

## 3. Authenticate with API Key

Your API key authenticates your connection to MPLAB ML. The `getpass` function hides your key as you type it for security.

**Need an API key?** See the [API key creation guide](../docs/getting-started/creating-an-api-key.md).

```python
# Secure API key input (characters will be hidden)
api_key = getpass("Enter your MPLAB ML API Key: ")

# Initialize client
client = Client(api_key=api_key)

print("✓ Connected to MPLAB ML!")
```

---
## 4. Explore Available SDK Functions

The MPLAB ML SDK provides functions for:
- Project management
- Data upload and management
- Pipeline and query operations
- Model and KnowledgePack operations

Let's see what's available:

```python
# Get all public methods
available_functions = [attr for attr in dir(client) if not attr.startswith('_')]

# Categorize functions
categories = {
    'Project Management': [],
    'Data Management': [],
    'Pipeline & Query': [],
    'Model & KnowledgePack': [],
    'Other': []
}

for func in available_functions:
    if 'project' in func.lower():
        categories['Project Management'].append(func)
    elif any(kw in func.lower() for kw in ['data', 'capture', 'segment', 'upload']):
        categories['Data Management'].append(func)
    elif any(kw in func.lower() for kw in ['pipeline', 'query']):
        categories['Pipeline & Query'].append(func)
    elif any(kw in func.lower() for kw in ['knowledgepack', 'model']):
        categories['Model & KnowledgePack'].append(func)
    else:
        categories['Other'].append(func)

# Display categorized functions
print("MPLAB ML SDK Functions by Category")
print("=" * 70)

for category, funcs in categories.items():
    if funcs:
        print(f"\n{category} ({len(funcs)} functions):")
        print("-" * 40)
        for func in sorted(funcs):
            print(f"  • {func}")
```

### Key Functions Reference

**Project Management:**
- `list_projects()` - View all your projects
- `mplabml_project_download()` - Download complete project
- `mplabml_project_upload()` - Upload project with data and labels

**Data Management:**
- `list_captures()` - View data captures in a project
- `upload_dataframe()` - Upload data from pandas DataFrame
- `list_segments()` - View labeled segments

**Pipeline & Query:**
- `list_pipelines()` - View feature extraction pipelines
- `list_queries()` - View data queries

**Model & KnowledgePack:**
- `list_models()` - View trained models
- `list_knowledgepacks()` - View generated Knowledge Packs

## 5. List Your Projects

Let's see what projects are in your MPLAB ML account:

```python
# Get list of all projects
projects_df = client.list_projects()

print(f"Total Projects: {len(projects_df)}\n")

# Display projects
if len(projects_df) > 0:
    display(projects_df)
else:
    print("No projects found. Let's create one!")
    print("See the data labeling notebooks to learn how to upload your first project.")
```

### Understanding the Projects DataFrame

Each project has:
- **Name** - Project identifier
- **Last Modified** - Most recent update timestamp
- **Created** - Project creation date
- **UUID** - Unique project identifier
